```python
df= df.drop(columns = ['id_songs','name_artists'])
```

```python
df.columns
```

```python
df['name_song'] =df['name_song'].str.strip().str.title()
```

```python
features = [
    "danceability","energy","loudness","speechiness","acousticness",
    "instrumentalness","liveness","valence","tempo","duration_ms"
]

x = df[features].copy()
```

```python
from sklearn.preprocessing import StandardScaler

X = df[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

sse = []
silhouette_scores = []
K = range(2, 8)

for k in K:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    model.fit(X_scaled)
    sse.append(model.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, model.labels_))
    
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(K, sse, '-o')
plt.title("Elbow Method (SSE vs K)")
plt.xlabel("k")
plt.ylabel("SSE")

plt.subplot(1,2,2)
plt.plot(K, silhouette_scores, '-o')
plt.title("Silhouette Score vs K")
plt.xlabel("k")
plt.ylabel("Silhouette Score")

plt.show()
```

```python
max_score = silhouette_scores.index(max(silhouette_scores))
n_clusters= K[max_score]
print("Total clusters",n_clusters)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

kmeans = KMeans(n_clusters=3, random_state=42, n_init=20)
labels_kmeans = kmeans.fit_predict(X_scaled)
df['cluster'] = labels_kmeans

# Evaluation Metrics (KMeans)
silhouette = silhouette_score(X_scaled, labels_kmeans)
db_index = davies_bouldin_score(X_scaled, labels_kmeans)

print("Silhouette Score (KMeans):", silhouette)
print("Davies-Bouldin Index (KMeans):", db_index)
```

```python
for c in sorted(df['cluster'].unique()):
    plt.figure(figsize=(8,6))
    subset = df[df['cluster'] == c][features]
    sns.heatmap(subset.corr(), annot=False, cmap='coolwarm')
    plt.title(f"Feature Correlation Heatmap - Cluster {c}")
    plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=labels_kmeans, cmap="tab10", s=10)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("PCA Visualization of Clusters (KMeans)")
plt.colorbar(label="Cluster")
plt.show()
```

```python
cluster_names = {
    0: "Party",
    1: "Calm acoustic",
    2: "Pop"
}
df['cluster_label'] = df['cluster'].map(cluster_names)
# Cluster sizes
cluster_sizes = df['cluster'].value_counts().sort_index()
print("\nCluster Sizes:")
cluster_sizes
```

```python
final_summary = (
    df.groupby(['cluster','cluster_label'])
      [features]
      .mean()
      .round(3)
      .reset_index()
)
print("\n Final Cluster Summary ")
final_summary
```

```python
import joblib
joblib.dump(features, "feature_col.pkl")

df.to_csv("amazon_music_clusters.csv", index=False)
print("\n Cluster labels added to dataframe and saved as 'amazon_music_clusters.csv'")
final_summary.to_csv("amazon_music_fs.csv", index=False)
```
